### 12/8 Music21

```python
import pandas as pd

df = pd.read_pickle("dataframes/final_version/music21_128.pkl")
```

```python
df
```

In 12/8 I'm considering first a division by 2, then a division by 2 and then by 3
    1. rule 3
    2. rule 8
    3. rule 14
    
So let's aggregate the rules for all the dataframe

```python
sum_rule= {}
for i in range(df.shape[0]):
    rule_list = df.loc[i]['used_rules']
    sum_rule = { k: sum_rule.get(k, 0) + rule_list.get(k, 0) for k in set(sum_rule) | set(rule_list) } #"sum" the two dictionaries

print(sum_rule)
```

```python
#import the vertical grammar

import pickle

with open('grammars_test/json/vert_grammar1.pkl', 'rb') as outfile:  
    vert_grammar = pickle.load(outfile)
    
# reset all weights to 0

for r in vert_grammar["rules"]:
    r["weight"] = 0

# add for each element in sum_rule

for r in sum_rule:
    vert_grammar["rules"][int(r)]["weight"] += sum_rule[r]
    
# find the total usage for the elements with the same head
head_usage={}
for r in vert_grammar["rules"]:
    if r["head"] in head_usage:
        head_usage[r["head"]] += r["weight"]
    else:
        head_usage[r["head"]] =  r["weight"]
    
#now normalize for each head
for r in vert_grammar["rules"]:
    if head_usage[r["head"]] != 0:
        r["weight"] = r["weight"]/head_usage[r["head"]]
```

```python
vert_grammar["rules"]
```

```python
# total number of measures, incorrect sum measure and failed trees
wrong_sum_total = df['wrong_sum_measures'].sum()
failed_total = df['failed_trees'].sum()
measures = df['measures'].sum()
    
print("Number of measures: {}".format(measures) )
print("Number of wrong sum measures: {}".format(wrong_sum_total) )
print("Number of failed trees measures: {}".format(failed_total) )
```

The first division by 2 was performed 74% 

The second division by 2 was performed 60%

The third division by 3 was performed 71%
